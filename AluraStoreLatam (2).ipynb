### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda2.head()
```

#1. Análisis de facturación



```python
#Ingreso total de cada tienda
total_tienda1 = tienda1['Precio'].sum()
total_tienda2 = tienda2['Precio'].sum()
total_tienda3 = tienda3['Precio'].sum()
total_tienda4 = tienda4['Precio'].sum()
```

```python
# Mostrar los los ingresos totales de cada tienda
print(f"Ingreso total Tienda 1: {total_tienda1}")
print(f"Ingreso total Tienda 2: {total_tienda2}")
print(f"Ingreso total Tienda 3: {total_tienda3}")
print(f"Ingreso total Tienda 4: {total_tienda4}")
```

# 2. Ventas por categoría

```python
#Agrupando cantidad de productos vendidos por categoría en Tienda 1
categorias_t1 = tienda1.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
print("Tienda 1 - Categorías más populares:\n", categorias_t1)
```

```python
#Agrupando cantidad de productos vendidos por categoría en Tienda 2
categorias_t2 = tienda2.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
print("Tienda 2 - Categorías más populares:\n", categorias_t2)
```

```python
#Agrupando cantidad de productos vendidos por categoría en Tienda 3
categorias_t3 = tienda3.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
print("Tienda 3 - Categorías más populares:\n", categorias_t3)
```

```python
#Agrupando cantidad de productos vendidos por categoría en Tienda 4
categorias_t4 = tienda4.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
print("Tienda 4 - Categorías más populares:\n", categorias_t4)
```

# 3. Calificación promedio de la tienda


```python
#Satisfacción del cliente con los productos vendidos por cada tienda
satisfaccion_tienda1 = tienda1['Calificación'].mean()
satisfaccion_tienda2 = tienda2['Calificación'].mean()
satisfaccion_tienda3 = tienda3['Calificación'].mean()
satisfaccion_tienda4 = tienda4['Calificación'].mean()
```

```python
# Mostrar satisfacción para cada tienda
print(f"Satisfacción - Tienda 1: {satisfaccion_tienda1:.2f}")
print(f"Satisfacción - Tienda 2: {satisfaccion_tienda2:.2f}")
print(f"Satisfacción - Tienda 3: {satisfaccion_tienda3:.2f}")
print(f"Satisfacción - Tienda 4: {satisfaccion_tienda4:.2f}")
```

# 4. Productos más y menos vendidos

```python
# Productos que fueron los más vendidos y los menos vendidos en cada tienda.
# Contar cuántas veces se vendió cada producto en cada tienda
ventas_t1 = tienda1['Producto'].value_counts()
ventas_t2 = tienda2['Producto'].value_counts()
ventas_t3 = tienda3['Producto'].value_counts()
ventas_t4 = tienda4['Producto'].value_counts()
```

```python
# Producto más vendido
print("Tienda 1 - Más vendido:\n", ventas_t1.idxmax(), "-", ventas_t1.max())
print("Tienda 2 - Más vendido:\n", ventas_t2.idxmax(), "-", ventas_t2.max())
print("Tienda 3 - Más vendido:\n", ventas_t3.idxmax(), "-", ventas_t3.max())
print("Tienda 4 - Más vendido:\n", ventas_t4.idxmax(), "-", ventas_t4.max())

# Producto menos vendido
print("Tienda 1 - Menos vendido:\n", ventas_t1.idxmin(), "-", ventas_t1.min())
print("Tienda 2 - Menos vendido:\n", ventas_t2.idxmin(), "-", ventas_t2.min())
print("Tienda 3 - Menos vendido:\n", ventas_t3.idxmin(), "-", ventas_t3.min())
print("Tienda 4 - Menos vendido:\n", ventas_t4.idxmin(), "-", ventas_t4.min())
```

# 5. Envío promedio por tienda

```python
# Costo de envío promedio para cada tienda.
costo_envio_tienda1 = tienda1['Costo de envío'].mean()
costo_envio_tienda2 = tienda2['Costo de envío'].mean()
costo_envio_tienda3 = tienda3['Costo de envío'].mean()
costo_envio_tienda4 = tienda4['Costo de envío'].mean()
```

```python
#Mostrar costo de envio por cada tienda.
print(f"Costo envio Tienda 1: {costo_envio_tienda1:.2f}")
print(f"Costo envio Tienda 2: {costo_envio_tienda2:.2f}")
print(f"Costo envio Tienda 3: {costo_envio_tienda3:.2f}")
print(f"Costo envio Tienda 4: {costo_envio_tienda4:.2f}")
```

#Visualización geográfica

```python
def graficar_geolocalizacion(tienda, nombre_tienda):
    plt.figure(figsize=(8, 6))
    plt.scatter(tienda['lon'], tienda['lat'], alpha=0.4, c='royalblue')
    plt.title(f'Distribución geográfica de ventas - {nombre_tienda}')
    plt.xlabel('Longitud')
    plt.ylabel('Latitud')
    plt.grid(True)
    plt.savefig(f'geo_{nombre_tienda.lower().replace(" ", "_")}.png')
    plt.show()

graficar_geolocalizacion(tienda1, 'Tienda 1')
graficar_geolocalizacion(tienda2, 'Tienda 2')
graficar_geolocalizacion(tienda3, 'Tienda 3')
graficar_geolocalizacion(tienda4, 'Tienda 4')
```

```python
import folium
from folium.plugins import HeatMap

def mapa_interactivo(tienda, nombre_tienda):
    # Punto central promedio
    centro = [tienda['lat'].mean(), tienda['lon'].mean()]

    mapa = folium.Map(location=centro, zoom_start=10)
    for _, fila in tienda.iterrows():
        folium.CircleMarker(
            location=[fila['lat'], fila['lon']],
            radius=2,
            color='blue',
            fill=True,
            fill_opacity=0.4
        ).add_to(mapa)

    return mapa

mapa_interactivo(tienda1, "Tienda 1")
```

```python
from folium.plugins import HeatMap

def mapa_calor(tienda, nombre_tienda):
    centro = [tienda['lat'].mean(), tienda['lon'].mean()]
    mapa = folium.Map(location=centro, zoom_start=10)

    heat_data = [[row['lat'], row['lon']] for index, row in tienda.iterrows()]
    HeatMap(heat_data).add_to(mapa)

    return mapa

mapa_calor(tienda1, "Tienda 1")
```
